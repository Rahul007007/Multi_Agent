```python
try:
    import os
    from crewai import Agent, Task, Crew, Process
except:
    import os
    from crewai import Agent, Task, Crew, Process
```

```python
from langchain.tools import DuckDuckGoSearchRun
from langchain.agents import Tool
from crewai_tools import ScrapeWebsiteTool
from crewai_tools import CodeInterpreterTool
from gnews import GNews
google_news = GNews()
google_news.period = '3d' # News from last 1 day
google_news.max_results = 7 # number of responses across a keyword
google_news.exclude_websites = ['yahoo.com', 'cnn.com', 'msn.con'] # Exclude news from specific website i.e Yahoo.com and CNN.com
#use date range if required
#google_news.start_date = (2023, 1, 1) # Search from 1st Jan 2023
#google_news.end_date = (2023, 4, 1) # Search until 1st April 2023

def search_recent_events(input_text):
    news_by_keyword = google_news.get_news(input_text)
    return news_by_keyword

# To enable scrapping any website it finds during it's execution
scrape_tool = ScrapeWebsiteTool()

search = DuckDuckGoSearchRun(max_results=1)

duck_search_tool = Tool(
        name = "Internet Search",
        func=search.run,
        description="useful when you need to search the internet for recent answers",
    )
code_tool = CodeInterpreterTool()

search_news_tool = Tool(
        name = "Internet Search for News",
        func=search_recent_events,
        description="useful when you need to search the internet for recent answers",
    )
```

```python
max_iter=5
```

```python
pro_coder = Agent(
    role="Skilled Coder",
    goal="Writes the best error-free, optimised code for any question given.",
    backstory="You're an expert coder who has knowledge about datastructures and programming in general.",
    allow_delegation=False,
    llm=code_llm,
    tools=[code_tool],
    max_iter=max_iter
)

scraper = Agent(
    role="Website Scraper",
    goal="Scrapes any website to obtain meaningful information from the website.",
    backstory="You're an expert at scraping websites and summarizing the important points mentioned."
            "You can also get URL suggestions for scraping from the Researcher for better information extraction.",
    allow_delegation=False,
    llm=groq_small_llm,
    tools=[scrape_tool],
    max_iter=max_iter
)

mathematician = Agent(
    role="Skilled Mathematican",
    goal="Solves the toughest of the math problems thrown at you.",
    backstory="You're an expert mathemetician and scientist who knows everything about maths and can solve any math problem given",
    allow_delegation=False,
    llm=math_llm,
    max_iter=max_iter
)

researcher = Agent(
    role="Researcher",
    goal="Conduct thorough research and analysis on the topic given. You can access the internet whenever you want to for better research.",
    backstory="You're an expert researcher in general who always gives the best research on the given topic."
                " You collect information that helps the "
                "audience learn something "
                "and make informed decisions. "
                "Your work is the basis for "
                "the Senior Writer to write an article on this topic."
                "Additionally you can also research on the topic and suggest website URLs to the Website Scraper "
                "for extracting additional information from the websites for better task execution",
    allow_delegation=False,
    llm=groq_small_llm,
    tools=[duck_search_tool, search_news_tool],
    max_iter=max_iter
)

writer = Agent(
    role="Senior Writer",
    goal="Write insightful, compelling and factually accurate opinion piece about the topic given",
    backstory="You're a highly adaptable senior writer who understands the context provided by the Researcher or the Website scraper and writes the best blogs."
                "You base your writing on the work of "
                "the Researcher or the Website Scraper, who provides an outline "
                "and relevant context about the topic.",
    allow_delegation=False,
    llm=groq_small_llm,
    max_iter=max_iter
)

# Define your task
task = Task(
    description="Perform this task: {input}",
    expected_output="Give me a short summary incase of website scraping, research topics, etc otherwise direct answers for math or coding questions",
)

# Define the manager agent
manager = Agent(
    role="Project Manager",
    goal="Efficiently manage the crew and ensure high-quality task completion",
    backstory="You're an experienced project manager, skilled in overseeing complex projects and guiding teams to success. Your role is to coordinate the efforts of the crew members, ensuring that each task is completed on time and to the highest standard.",
    allow_delegation=True,
    llm=groq_main_llm
)

# Instantiate your crew with a custom manager
crew = Crew(
    agents=[scraper, pro_coder, researcher, writer, mathematician],
    tasks=[task],
    manager_agent=manager,
    process=Process.hierarchical,
    verbose=1,
    cache=True,
    memory=False,
    embedder={
                "provider": "openai",
                "config":{
                        "model": 'text-embedding-3-small',
                }
        }
)


# Start the crew's work
output=crew.kickoff(inputs={"input": "What is about this Microsoft outage today?"})
```
